```python
from collections import Counter
cnt = Counter("abracadabra")
start = Counter(bike_share['start station id'])
print max(start)
end = Counter(bike_share['end station id'])
print max(end)
import operator

print max(start.iteritems(), key=operator.itemgetter(1))[0]
print max(end.iteritems(), key=operator.itemgetter(1))[0]
print start[293]
print end[293]

net = { k: end.get(k, 0) - start.get(k, 0) for k in set(x) | set(y) }
print max(net.iteritems(), key=operator.itemgetter(1))[0]
print min(net.iteritems(), key=operator.itemgetter(1))[0]

print net[531]
print net[517]

```

```python
plt.hist(net.values())
plt.title('Net Flow in February 2014')
plt.xlabel('Net Flow of Bikes')
plt.ylabel('Number of Stations')
```

For most stations, no transfer needed, few dramatic sources, rather than sinks
